```python
"""
Connect 4 Game
By: Shira Drechsler
Date: Aug 20, 2020
"""

rows = [[], [], [], [], [], []]

headers='1 2 3 4 5 6 7'
for row in range(6):
    for col in range(7): 
        rows[row].append('â€¢')
    

r1 = ' '.join(rows[0])
r2 = ' '.join(rows[1])
r4 = ' '.join(rows[3])
r5 = ' '.join(rows[4])
r3 = ' '.join(rows[2])
r6 = ' '.join(rows[5])

print(headers)
print(r1)
print(r2)
print(r3)
print(r4)
print(r5)
print(r6)
   
playerO = input('Player O, tell me your name: ')
playerX = input('Player X, what should I call you? ')
print('Okay, ' + playerO + ' and ' + playerX + "! Let's start! ")

winner = False
fullboard = False
player = 1
toprow = {1:6, 2:6, 3:6, 4:6, 5:6, 6:6, 7:6}

while not winner and not fullboard:
        goodcol = False
        if player==1: 
             symbol='X'
             playerturn = playerX
        else: 
             symbol = 'O'
             playerturn = playerO
             
        while not goodcol:  
            col = int(input(playerturn + ', pick a column:'))
            while col not in [1,2,3,4,5,6,7]:
                col = int(input('Sorry, there is no column ' + str(col) + '. Pick again: '))
            
            if toprow[col] == 0:
               print('That column is full, pick a different column.')
            else:
                goodcol = True
                toprow[col]=toprow[col]-1
            
        rows[toprow[col]][col-1] = symbol
         
        r1 = ' '.join(rows[0])
        r2 = ' '.join(rows[1])
        r3 = ' '.join(rows[2])
        r4 = ' '.join(rows[3])
        r5 = ' '.join(rows[4])
        r6 = ' '.join(rows[5])

        print(headers)
        print(r1)
        print(r2)
        print(r3)
        print(r4)
        print(r5)
        print(r6)
        
         
        # check if board is full and update 
        same_symbol = True 
        counter1 = 0
        cur_row = toprow[col]+2  #start one below the current chip
        cur_col = col
        while same_symbol == True and cur_row <= 6: 
              if rows[cur_row-1][cur_col-1]==symbol:
                 counter1 += 1
                 same_symbol = True 
              else:
                 same_symbol = False 
              cur_row += 1 
        
        if counter1 >=3 :
              winner=True 
        
        counterright = 0
        cur_row = toprow[col]+1
        cur_col = col + 1
        same_symbol = True 
        while same_symbol == True and cur_col <= 7:
              if rows[cur_row-1][cur_col-1]==symbol:
                 counterright += 1
                 same_symbol = True
              else:
                  same_symbol = False
              cur_col += 1
              
        counterleft = 0  
        cur_row = toprow[col] + 1
        cur_col = col - 1
        same_symbol = True
        while same_symbol == True and cur_col >= 1:
              if rows[cur_row-1][cur_col-1]==symbol:
                 counterleft += 1
                 same_symbol = True
              else:
                  same_symbol = False
              cur_col -= 1
        sidesum = counterright + counterleft
        if sidesum >= 3:
           winner = True
           
        counterdownright = 0
        cur_row = toprow[col] + 2 
        cur_col = col + 1
        same_symbol = True
        while same_symbol == True and cur_col <= 7 and cur_row <= 6:
              if rows[cur_row-1][cur_col-1] == symbol:
                 counterdownright += 1
                 same_symbol = True
              else:
                  same_symbol = False
              cur_col += 1
              cur_row += 1
              
        counterupleft = 0
        cur_row = toprow[col] 
        cur_col = col - 1
        same_symbol = True
        while same_symbol == True and cur_col >=1  and cur_row >=1:
              if rows[cur_row-1][cur_col-1] == symbol:
                 counterupleft += 1
                 same_symbol = True
              else:
                  same_symbol = False
              cur_col -= 1
              cur_row -= 1
              
        diag1sum = counterdownright + counterupleft
        if diag1sum >= 3 :
           winner = True
          
        counterdownleft = 0
        cur_row = toprow[col] + 2 
        cur_col = col -1
        same_symbol = True
        while same_symbol == True and cur_col >= 1 and cur_row <= 6:
              if rows[cur_row-1][cur_col-1] == symbol:
                 counterdownleft += 1
                 same_symbol = True
              else:
                  same_symbol = False
              cur_col -= 1
              cur_row += 1
          
        counterupright = 0
        cur_row = toprow[col] 
        cur_col = col + 1
        same_symbol = True
        while same_symbol == True and cur_col <=7  and cur_row >=1:
              if rows[cur_row-1][cur_col-1] == symbol:
                 counterupright += 1
                 same_symbol = True
              else:
                  same_symbol = False
              cur_col += 1
              cur_row -= 1  
            
        diag2sum = counterdownleft + counterupright
        if diag2sum >= 3 :
           winner = True
          
        if not winner :
            if player == 1: 
                player =2
            else: 
                player =1    
           
        topsum = 0
        for i in range(1, 8):
            topsum += toprow[i]
            
        if topsum == 0:
           fullboard = True
           
if winner==True:
    print('\n Connect 4! ' + playerturn + ' wins!' )
else:
    print("The board's full. Too bad! Better luck next time!")        
           
```
